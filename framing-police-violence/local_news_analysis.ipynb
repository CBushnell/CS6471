```python
import pandas as pd
import numpy as np
import re
from IPython.display import display, HTML
```

```python
articles = pd.read_csv('./data/prepared/shootings/all_dated.csv')
articles.head(2)
```

```python
sources = pd.read_csv('./resources/news_sources/local_news_dataset.csv')
sources
```

```python
local_articles = articles.__deepcopy__()
local_articles["news_city"] = pd.NA
sites = [re.compile(".*" + x + ".*") for x in sources['Site'].tolist()]
cities = [x for x in sources['City']]
urls = articles['url']
local_cities = pd.NA
locals = np.full(len(local_articles), False)
for i in range(len(urls)):
    for j, site in enumerate(sites):
        if site.match(urls[i]):
            locals[i] = True
            local_articles.loc[i, "news_city"] = cities[j]
```

```python
local_articles = local_articles.loc[locals]
```
